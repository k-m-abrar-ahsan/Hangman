```python
words=['MATH', 'ALGEBRA', 'UNIVERSITY','BANGLADESH','COMPUTER','MEMORIES','DANGER'] #This list need more words to make the game more fun
def board(idx):
  for i in idx:
    lst[i]=x
  strg=''
  for i in lst:
    strg+=i
  print(strg)
```

```python
  
graphic = [
        """
            +-------+
            |
            |
            | 
            |
            |
         ==============
        """
            ,
        """
            +-------+
            |       |
            |       0
            | 
            |
            |
         ==============
        """
            ,
        """
            +-------+
            |       |
            |       0
            |       |
            |
            |
         ==============
        """
            ,
        """
            +-------+
            |       |
            |       0
            |      -|
            |
            |
         ==============
        """
            ,
        """
            +-------+
            |       |
            |       0
            |      -|-
            |
            |
         ==============
        """
            ,
        """
            +-------+
            |       |
            |       0
            |      -|-
            |      /
            |
         ==============
        """
            ,
        """
            +-------+
            |       |
            |       0
            |      -|-
            |      / \
            |
         ==============
        """]
```

```python

import random
import re
counter=0
n=0
chances=len(graphic)
word=random.choice(words)


lst=[' _ ']*len(word)

print('Word is of',len(word),'letters')
while True:
  
  x=input().upper()
  if x in word:
    print('Guess is correct')
    counter+= word.count(x)
    idx=[i.start() for i in re.finditer(x, word)]
    board(idx)
  else:
    n+=1
    print(graphic[n-1])
    print('Guess is not correct')

  print('Chances left:', chances-n)
  if chances - n ==0 :
    print('Over')
    print('Word:',word)
    break
  if len(word) - counter == 0:
    print('Win')
    print('Word:',word)
    break
```
